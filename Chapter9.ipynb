## Figures and Subplots

The *Figure object* is matplotlib's core, within it you can find differents plots named *Subplots* (These are other objects commonly named axes). <br>
Add these subplots as follows:


```python
fig=plt.figure(figsize=(2,2))
ax1=fig.add_subplot()
```

These plot axis objects have many methods that create different types of plots. It is preferable to use them over the top-level functions as plt.plot() (*Because the top level functions could overwrite your last figure used*)

The most used way to create a figure and its subplots is using the plt.subplots() function:

```python
fig,axes=plt.subplots(2,2)
```

Then you can select any of the subplots using indexing over the array axes:

```python
axes[0,1].plot([0.2,0.4])
fig
```
